# Apple Watch and the Prime Number Running Protocol

## October 2020 Run Durations That End In Prime Number Minutes

October 2020 running protocol: 

1. No runs shorter than 29 minutes.


2. The duration in minutes should be a prime number. My body and legs play a major role in determining which prime number to choose.


3. 10 seconds to stop my Apple Watch timer once my run duration hits the desired prime number minutes. Otherwise I have to continue running until I reach the next prime number. 
 

### Setup

```python
import pathlib
```

```python
import pandas as pd
```

```python
home = pathlib.Path.home()
```

```python
study_path = f"{home}/small-data/study/apple-watch-health-tracking/story-data"
```

```python
runs = pd.read_csv(f"{study_path}/apple_watch_runs_prime_202010.csv",parse_dates=['date'])
```

### Frequency Distribution of Runs With Prime Number Durations

```python
oct2020_runs = runs[runs['year and month'] == '2020-10']
```

```python
grouped_primes = oct2020_runs.groupby('Prime Number Minutes').count()
```

```python
barplot = grouped_primes.plot.bar(y="miles", 
                                  legend=False,
                                  color=['blue', 'orange', 'green', 'red', 'purple'],
                                  title = "Frequency Distribution of October 2020 Runs in Prime Number Minutes",
                                  figsize=(11, 7), rot=0, grid=True)
```

```python
barplot.get_figure().savefig(f"{study_path}/oct2020_prime_frequency")
```

### Prime Number Runs Across October 2020

```python
runs_plot = oct2020_runs.plot.line(y=['Prime Number Minutes'],
                           x='date',
                           title='Prime Number Runs, October 2020', 
                           color = 'orange',
                           markevery=1, markersize=14, style='.-',
                           ylim=(25,59), grid=True,
                           figsize=(13, 11), rot=0)
```

```python
runs_plot.get_figure().savefig(f"{study_path}/oct2020_prime_lineplot")
```
