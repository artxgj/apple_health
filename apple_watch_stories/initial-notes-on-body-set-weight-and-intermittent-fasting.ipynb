# A Look At Body Set Weight Reset and Intermittent Fasting

## Preliminary Notes

This is an initial attempt to grasp the concept of [body set weight] (https://medium.com/@drjasonfung/controlling-the-bodys-fat-thermometer-12e2e69e94dd) and intermittent fasting  using my own weight data. For this data report, I define A-L as a set of body set weight classes. Each class has a range of 5 lbs. A has the lowest range bounds while L has the highest range bounds. A change in cbody set class signifies a resetting of the body set weight.

Between September and November 2020, a quasi intermittent fasting took place. I estimate more than 80% of this period involved a 10-12 hour difference between dinner and the following day's first meals. In November, there is a concerted effort to break the fast after 12-16 hours. An idea has formed to turn December 2020 into a running offseason (there might be some very light running for some days) and doing an intermittent fast to see the effects on my weight. 

### Setup

```python
import pathlib
```

```python
import pandas as pd
```

```python
home = pathlib.Path.home()
```

```python
study_path = f"{home}/small-data/study/apple-watch-health-tracking/story-data"
```

```python
bodymass_divisions = pd.read_csv(f"{study_path}/bodymass-divisions.csv",parse_dates=['date'])
```

### Frequency Distribution of 2020 Body Set Weight Classes

```python
weights_2020 = bodymass_divisions[(bodymass_divisions['date'] > '2019-12-31') & (bodymass_divisions['date'] < '2021-01-01')]
```

```python
grouped_body_set_weight_classes = weights_2020.groupby('Body Set Weight Group').count()
```

```python
barplot = grouped_body_set_weight_classes.plot.bar(y="Body Set Weight Minimum", 
                                  legend=False,
                                  color=['blue', 'orange', 'green', 'red', 'purple'],
                                  title = "2020 Frequency Distribution of Weight Based On Boxing Weight Classes (WBA and WBC)",
                                  figsize=(11, 7), rot=0, grid=True)

barplot.set_ylabel("Number of Days")
```

```python
barplot.get_figure().savefig(f"{study_path}/body_set_weight_classes_barplot_2020")
```

### My 2020 Boxing Weight Classes (WBA/WBC)

```python
weights_2020_plot =  weights_2020.plot.line(y=['Body Set Weight Minimum'],
                           x='date',
                           title='2020: Moving Down Weight Classes (WBA and WBC)', 
                           color = 'orange',
                           markevery=3, markersize=2, style='.-',
                           ylim=(0,70), grid=True,
                           figsize=(17, 11), rot=0)

weights_2020_plot.set_ylabel("Weight Class Minimum Weight (lbs)")
```

```python
weights_2020_plot.get_figure().savefig(f"{study_path}/body_set_weight_classes_lineplot_2020")
```
