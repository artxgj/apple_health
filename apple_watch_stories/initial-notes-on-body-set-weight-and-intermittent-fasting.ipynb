I define body set weight reset as changing weight classes and staying in the new weight class for at least 10 consecutive days.

### Frequency Distribution of 2020 Body Set Weight Classes

```python
weights_2020 = bodymass_divisions[(bodymass_divisions['date'] > '2019-12-31') & (bodymass_divisions['date'] < '2021-01-01')]
```

```python
grouped_body_set_weight_classes = weights_2020.groupby('Body Set Weight Group').count()
```

```python
barplot = grouped_body_set_weight_classes.plot.bar(y="Body Set Weight Minimum", 
                                  legend=False,
                                  color=['blue', 'orange', 'green', 'red', 'purple'],
                                  title = "2020 Frequency Distribution of Body Weight Classes",
                                  figsize=(11, 7), rot=0, grid=True)


barplot.set_ylabel("Number of Days In Weight Class")
```

```python
barplot.get_figure().savefig(f"{study_path}/body_set_weight_classes_barplot_2020")
```

##### Move More and Eat Less #####

My experience has shown me that losing the first 10 lbs is easy. My fast way of achieving this is to eat less and move a little more. This method worked again in 2020 as I went from weight class I down to G within 6 weeks. I __do not count__ calories. I just eyeball the portion size of my meals. Move More for me means running and walking.

The second 10 lbs were harder to lose. My body set weight enjoyed long stays at the G and F classes. I ran and walked and watched my meal-portions from March to August. Near the middle of August, I reduced carbs intake from rice or pasta, and increased my weekly vegetables portions. August began the eggplant and bittermelon experiment. Every major meal included egpplant and/or bittermelon; I have two major meals a day for the majority of 2020. 


##### Apple Watch Exercise Minutes #####

The Apple Watch has a workout app that lets people record their workouts like outdoor run, outdoor walk, swimming, etc. Each workout duration is automatically added to the daily activity exercise bucket. In addition, the watch can detect movements and any detection of brisk activity over some duration is counted as exercise. One doesn't have to turn on the workout app to get exercise minutes. A leisurely walk for me or pushups over a minute can count as exercise to my Apple Watch.

##### Quasi Intermittent Fasting #####

From September to October, my body set weight oscillated between classes E and D. During these two months, I observed quasi intermittent fasting (10-12 hours between dinner and the following day's first meal). I also eliminated walking from my daily exercise and focused only on running. 

My body set weight stayed consistently in weight class D in the latter half of October. 

##### Intermittent Fasting 
Beginning in November, I began an experiment of a longer intermittent fast (15-16 hours between dinner and the first meal of the following day). I continue to run in November but reduced my daily average duration. I have been surprised by the results of reduced exercise time and intermittent fasting after 12 days. However, I do need more data. 

A future experiment with intermittent fasting will be one with very less than 30 minutes of exercise. I am curious to see how effective intermittent fasting is at maintaining weight when I don't move as much.
