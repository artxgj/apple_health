# A Look At Body Set Weight Reset and Intermittent Fasting

## Preliminary Notes

This is an initial attempt to grasp the concept of [body set weight](https://medium.com/@drjasonfung/controlling-the-bodys-fat-thermometer-12e2e69e94dd) and intermittent fasting  using my own weight data. I borrow boxing's weight classes and define my own classes of weight ranges. The Each weight class represents a body set weight. My weight classes span 70 pounds and each class has a range of 5 pounds. These weight ranges are not absolute weights. They just show an approximate increase or decrease of five pounds when I move from one weight class to another. 

I define body set weight reset as changing weight classes and staying in the new weight class for at least 10 consecutive days.

### Setup

```python
import pathlib
```

```python
import pandas as pd
```

```python
home = pathlib.Path.home()
```

```python
study_path = f"{home}/small-data/study/apple-watch-health-tracking/story-data"
```

```python
bodymass_divisions = pd.read_csv(f"{study_path}/bodymass-divisions.csv",parse_dates=['date'])
```

### Frequency Distribution of 2020 Body Set Weight Classes

```python
weights_2020 = bodymass_divisions[(bodymass_divisions['date'] > '2019-12-31') & (bodymass_divisions['date'] < '2021-01-01')]
```

```python
grouped_body_set_weight_classes = weights_2020.groupby('Body Set Weight Group').count()
```

```python
barplot = grouped_body_set_weight_classes.plot.bar(y="Body Set Weight Minimum", 
                                  legend=False,
                                  color=['blue', 'orange', 'green', 'red', 'purple'],
                                  title = "2020 Frequency Distribution of Body Weight Classes",
                                  figsize=(11, 7), rot=0, grid=True)

barplot.set_ylabel("Number of Days In Weight Class")
```

```python
barplot.get_figure().savefig(f"{study_path}/body_set_weight_classes_barplot_2020")
```

### Changes in Body Set Weight (Weight Classes)

```python
weights_2020_plot =  weights_2020.plot.line(y=['Body Set Weight Minimum'],
                           x='date',
                           title='2020: Changes in Weight Classes', 
                           color = 'orange',
                           markevery=3, markersize=2, style='.-',
                           ylim=(0,70), grid=True,
                           figsize=(13, 7), rot=0)

weights_2020_plot.set_ylabel("Weight Class Minimum Weight (lbs)")
```

| Weight class | Range (lbs) | Consistent Body Set Weight's First Month | Last Month of Body Set Weight |

My experience has shown me that losing the first 10 lbs is easy. My fast way of achieving this is to eat less and move a little more. This method worked again in 2020 as I went from weight class I down to G within 6 weeks.

The second 10 lbs were harder to lose. My body set weight enjoyed long stays at the G and F classes. I ran and walked and watched my meal-portions from March to August.

From September to October, my body set weight oscillated between classes E and D. During these two months, I observed quasi intermittent fasting (10-12 hours between dinner and the following day's first meal). I also eliminated walking from my daily exercise and focused only on running. 

My body set weight stayed consistently in weight class D in the latter half of October. Beginning in November, I began an experiment of a longer intermittent fast (15-16 hours between dinner and the first meal of the following day). I continue to run in November. 

A future experiment with intermittent fasting will be one without any running for a period of time. I am curious to see how effective intermittent fasting is at maintaining weight when I don't exercise.
