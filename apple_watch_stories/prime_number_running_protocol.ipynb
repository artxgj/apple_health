# Apple Watch and the Prime Number Running Protocol

## Run Durations That End In Prime Number Minutes

After putting in more running miles this year than I had in the past few years combined, I needed to find a way to stay motivated and to have some fun. I came up with this running protocol for September 2020: 

* No runs shorter than 20 minutes.


* The duration in minutes should be a prime number. 


* 10 seconds to stop my Apple Watch timer once my run duration hits the desired prime number minutes. Otherwise I have to continue running until I reach the next prime number. 

### Setup

```python
import pathlib
```

```python
import pandas as pd
```

```python
home = pathlib.Path.home()
```

```python
study_path = f"{home}/small-data/study/apple-watch-health-tracking/story-data"
```

```python
runs = pd.read_csv(f"{study_path}/apple_watch_runs_prime_202010.csv",parse_dates=['date'])
```

### Frequency Distribution of Runs With Prime Number Durations

```python
sept_runs = runs[runs['year and month'] == '2020-09']
```

```python
grouped_primes = sept_runs.groupby('Prime Number Minutes').count()
```

```python
barplot = grouped_primes.plot.bar(y="miles", 
                                  legend=False,
                                  color=['blue', 'orange', 'green', 'red', 'purple'],
                                  title = "Frequency Distribution of September 2020 Runs in Prime Number Minutes",
                                  figsize=(11, 7), rot=0, grid=True)
```

```python
barplot.get_figure().savefig(f"{study_path}/sep2020_prime_frequency")
```

### Prime Number Runs Across September 2020

```python
runs_plot = sept_runs.plot.line(y=['Prime Number Minutes'],
                           x='date',
                           title='Prime Number Runs, September 2020', 
                           color = 'orange',
                           markevery=1, markersize=14, style='.-',
                           ylim=(20,70), grid=True,
                           figsize=(17, 11), rot=0)
```

```python
runs_plot.get_figure().savefig(f"{study_path}/sep2020_prime_lineplot")
```
