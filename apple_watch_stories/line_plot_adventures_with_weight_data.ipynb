```python
ext_weight = pd.read_csv(f"{study_path}/extended-weight-attributes.csv", parse_dates=['date'])
```

### Relative Weight Changes Since Day 0

```python
ext_weight_days_plot_line = ext_weight.plot.line(title="Relative Weight Changes Since Day 0 over 910 days.", 
             x='day_ordinal', y='lb_delta_since_start',
             figsize=(17, 11), style='.-', 
             markevery=1, markersize=14, markerfacecolor='orange')
```

### Relative Weight Changes Since Day 0 Over Intervals

```python
accidentally_delightful_plot= ext_weight.plot.line(title="Relative Weight Changes Since Day 0 Over 25 intervals", 
             x='interval_ordinal', y='lb_delta_since_start',
             figsize=(17, 11), style='.-', 
             markevery=1, markersize=14, markerfacecolor='orange')
```

Each **interval** represents a month, but there can be more one than month between intervals. In addition, each **interval** does not necessarily start on the first of the month.

### Weight Changes Between Weigh-ins.

```python
ext_weight_days_plot_line = ext_weight.plot.line(title="Weight-changes between weigh-ins over a period of 910 days", 
             x='day_ordinal', y='lb_delta_since_last',
             figsize=(17, 11), style='.-', 
             markevery=1, markersize=14, markerfacecolor='orange')
```

```python
ext_weight_intervals_plot_line =  ext_weight.plot.line(title="Weight-changes Between Weigh-ins over 25 Intervals", 
             x='interval_ordinal', y='lb_delta_since_last',
             figsize=(17, 11), style='.-', 
             markevery=1, markersize=14, markerfacecolor='orange')
```

Intervals with more than one weigh-in show the range of weight measurements.
