## Fun Activity: Exploring Line-Plotting Using Weight Data

## Plus A Foray into Box Plots

### Setup

```python
import pathlib
```

```python
import pandas as pd
```

```python
home = pathlib.Path.home()
```

```python
study_path = f"{home}/small-data/study/apple-watch-health-tracking/story-data"
```

```python
ext_weight = pd.read_csv(f"{study_path}/weights-intervals-study.csv", parse_dates=['Date'])
```

### Relative Weight Changes Since Day 0

```python
ext_weight_days_plot_line = ext_weight.plot.line(title="Relative Weight Changes Since Day 0 over 933 days", 
             x='Day', y='Weight Change Since First Day',
             figsize=(17, 11), style='.-', grid=True,
             markevery=1, markersize=14, markerfacecolor='orange')
```

### Relative Weight Changes Since First Day Over Intervals

```python
accidentally_delightful_plot= ext_weight.plot.line(
    title="Relative Weight Changes Since Day 0 Over 27 intervals", 
    x='Interval', y='Weight Change Since First Day',
    figsize=(17, 11), style='.-', grid=True,
    markevery=1, markersize=14, markerfacecolor='orange')
```

Each **interval** represents a month, but there can be more one than month between intervals. In addition, each **interval** does not necessarily start on the first of the month.

#### Box plot version of previous line plot

```python
boxplot_interval_weight_change_since_day1 = ext_weight.boxplot(
    by='Interval',
    column=["Weight Change Since First Day"], 
    figsize=(17, 11), 
    color="#4984b8")
```

### Weight Changes Between Weigh-ins.

```python
ext_weight_days_plot_line = ext_weight.plot.line(title="Weight Changes Between Weigh-ins Over 933 Days", 
             x='Day', y='Weight Change Since Last Weigh-in',
             figsize=(17, 11), style='.-', grid=True,
             markevery=1, markersize=14, markerfacecolor='orange')
```

```python
ext_weight_intervals_plot_line =  ext_weight.plot.line(title="Weight Changes Between Weigh-ins over 27 Intervals", 
             x='Interval', y='Weight Change Since Last Weigh-in',
             ylim=(-6, 10),                                          
             figsize=(17, 11), style='.-', grid=True,
             markevery=1, markersize=14, markerfacecolor='orange')
```

#### Box plot version

```python
boxplot_interval_weight_change_since_last_weighin = ext_weight.boxplot(
    by='Interval',
    column=["Weight Change Since Last Weigh-in"], 
    figsize=(17, 11), 
    color="#4984b8")
```
